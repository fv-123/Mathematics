Binomial Distribution

```python
# A stock has a 60% chance of going up each day. If you observe the stock for 5 days, what is the probability that it goes up exactly 3 times?

import math
n = 5
rate = 3/5
distribution=list()
for x in range(n+1):
    p=math.comb(n,x) * rate**x * (1-rate)**(n-x)
    distribution.append(p)
print(distribution)
```

Poisson Distribution

```python
#A trading algorithm detects an average of 4 arbitrage opportunities per day. What is the probability that exactly 2 arbitrage opportunities appear in a day?

lamb=4
distribution=list()
for x in range(10):
    p=(lamb**x * math.e**(-lamb))/math.factorial(x)
    distribution.append(p)
print(distribution)
```

Geometric Distribution

```python
#A trader makes a trade every day, and 10% of trades are profitable. What is the probability that the first profitable trade happens on the 4th day?

rate=0.1

distribution=list()
for x in range(1,10):
    p=(1-rate)**(x-1) * rate
    distribution.append(p)
print(distribution)
```

```python
# Cumulative Distribution Function for Geometric Distribution
#Probability of at least 1 success in until x period
cumulative=list()
for x in range(1,10):
    p = 1 - (1-rate)**x
    cumulative.append(p)
print(cumulative)
```

Monty Hall

```python
import random as rd

def the_game():
    trial=1000
    k_win,s_win=0,0
    for _ in range(trial):
        doors=[1,2,3]
        win=rd.choice(doors)
        
        choice=rd.choice(doors)
        
        if win==choice:
            after_choose=[win]
            after_choose.append(rd.choice([d for d in doors if d not in after_choose]))
        elif win!=choice:
            after_choose=[win,choice]
        
        keep_win=(choice==win)
        if keep_win:
            k_win+=1
            
        switch_choice= [d for d in after_choose if d!=choice][0]
        switch_win=(switch_choice==win)
        if switch_win:
            s_win+=1
    return f"If you dont change, you win {k_win} times. If you change, you win {s_win} times"
    
the_game()
```

TFT Shop

```python
import numpy as np

def simulate_tft_shop(trials, total_5_costs, remaining_kaisa, slots=5, level_9_prob=0.1):
    success_count = 0
    
    for _ in range(trials):
        for _ in range(slots):
            if np.random.rand() < level_9_prob:  # 10% chance of 5-cost
                chosen_champ = np.random.choice(total_5_costs)  # Pick from available 5-costs
                if chosen_champ == 'KaiSa':
                    success_count += 1
                    break  # No need to check further slots in this shop
    
    return success_count / trials  # Probability estimate

# Case 1: All other 5-costs available
total_5_costs_case1 = ['KaiSa'] * 1 + ['Other'] * 90  # 91 total remaining copies
prob_case1 = simulate_tft_shop(100000, total_5_costs_case1, 1)

# Case 2: Half of other 5-costs taken
total_5_costs_case2 = ['KaiSa'] * 1 + ['Other'] * 45  # 46 total remaining copies
prob_case2 = simulate_tft_shop(100000, total_5_costs_case2, 1)

print(f"Estimated probability of rolling KaiSa (all 5-costs available): {prob_case1:.4%}")
print(f"Estimated probability of rolling KaiSa (half of other 5-costs taken): {prob_case2:.4%}")
```
