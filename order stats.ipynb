```python
import numpy as np

# Parameters
n_samples = 1_000_000  # number of simulations

samples = np.random.uniform(0,1, size=(n_samples,3))

maxes = np.max(samples,axis=1)
mins = np.min(samples, axis=1)
median = np.median(samples, axis=1)

estimated_expectation_min = np.mean(mins)
estimated_expectation_max = np.mean(maxes)
estimated_expectation_med = np.mean(median)

print(f"Estimated E[min]: {estimated_expectation_min:.5f}")
print(f"Estimated E[max]: {estimated_expectation_max:.5f}")
print(f"Estimated E[median]: {estimated_expectation_med:.5f}")
```

```python
import sympy as sp
# CDF for the min (using complement)
y = sp.symbols('y')

cdf = 1 - (1 - y)**3

der_cdf = sp.diff(cdf, y)
print("PDF: ", der_cdf)

E = sp.integrate(y * der_cdf, (y, 0, 1))
print("expected value:", E)
```
