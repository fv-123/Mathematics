```python
# Adding derivative: Polynomial example
import sympy as sp

x = sp.Symbol('x')
f_x = sp.sympify(input("enter function: "))
sp.pprint(f_x)

def polynomial_der(f_x,x):
    # Convert to simpy expression
    f_sym=sp.sympify(f_x)
    terms = f_sym.as_ordered_terms()
    result = 0
    for term in terms:
        coeff, power = term.as_coeff_exponent(x)
        if power == 0: 
            d_term = 0
        else:
            d_term = power * coeff * x**(power - 1)
        result += d_term
    return result
polynomial_der(f_x,x)
```

```python
# Adding and Multiplication rule


import sympy as sp

x = sp.Symbol('x')
f_x = sp.sympify(input("enter function: "))
sp.pprint(f_x)

def der(func, x):
    if func.is_Atom:
        return 1 if func == x else 0
    if func.is_Add:
        return sum(der(arg, x) for arg in func.args)
    if func.is_Mul:
        terms = func.args
        result = 0
        for i in range(len(terms)):
            d_term = der(terms[i],x)
            others = (terms[j] for j in range(len(terms)) if j != i)
            rest = sp.Mul(*others)
            result += d_term * rest
        return result
    if func.is_Pow:
        base, exp = func.args
        if base == x and exp.is_Number:
            return exp * base ** (exp - 1)
        elif exp.is_Number:
            return exp * base ** (exp - 1) * der(base, x)
        else:
            return func * (der(exp, x) * sp.log(base) + exp * der(base, x) / base)
    if func.func == sp.log:
        u = func.args[0]
        return der(u,x) / u
    if func.func == sp.exp:
        u = func.args[0]
        return der(u,x) * sp.exp(u)
    if func.func == sp.sin:
        u = func.args[0]
        return der(u,x) * sp.cos(u)
    if func.func == sp.cos:
        u = func.args[0]
        return -der(u, x) * sp.sin(u)

first_der = der(f_x,x)
sec_der = der(first_der,x)

der_expanded = sp.expand(first_der)

der_collected = sp.collect(der_expanded, x)

print("first derivarive is: ")
sp.pprint(first_der)
print("second derivative is: ")
sp.pprint(sec_der)
```
